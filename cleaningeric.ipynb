```python
import pandas as pd
```

```python
df = pd.read_csv("adform2.csv")
```

```python
df['area_num'] = df['area_num'].str.replace('-', '')
```

```python
print(df['rent_act'].unique())

df.loc[df['rent_act'].str.contains("None|none|none available|non| nan|None vacant|Nil|No units available", na=False), "rent_act"] = "None"
df.loc[df['rent_act'].str.contains("Poor|Por|Poor to Fair|poor", na=False), "rent_act"] = "Poor"
df.loc[df['rent_act'].str.contains("Fair|fair|~fair~|Fair to good|fair|Fair only|Fair (a)|Only fair", na=False), "rent_act"] = "Fair"
df.loc[df['rent_act'].str.contains("Good|Good |good|Good  |Good- no vacancies|Good, by low income workers only| good|Very good|God|good - none available|good ", na=False), "rent_act"] = "Good"
df.loc[df['rent_act'].str.contains("Excellent|Xcellent|Excellent ", na=False), "rent_act"] = "Excellent"
df.loc[df['rent_act'].str.contains("All occupied|All filled|100% Filled|Filled up|all filled|Filled|Full|All filled up", na=False), "rent_act"] = "Filled"
df.loc[df['rent_act'].str.contains("-|- - - -|- - -|---------|--|-----|----------", na=False), "rent_act"] = "-"
df.loc[df['rent_act'].str.contains("90% filled|85 occupied|95% filled|few available|Few good vacancies|Almost none for rent|About filled", na=False), "rent_act"] = "Almost Full"
df.loc[df['rent_act'].str.contains("Light|light|Very light|very light|very light ", na=False), "rent_act"] = "Light"
```

```python
print(df['rentdemand1'].unique())

df.loc[df['rentdemand1'].str.contains("No|no |none|Nothing for rent|No|no |None|No rental properties|None-Not a rent area|Nont a rent area|No rentals|no rentals|Not a rental area|Not a rent area|Not rental area|non|no rents", na=False), "rentdemand1"] = "None"
df.loc[df['rentdemand1'].str.contains("Fair|fair|~good~ fair|Fair to good|Fair |Fairly good|Fair to poor|Fair only|fair to good|Fair-good|Fair-Good|Fair to Good|Only fair|fair ", na=False), "rentdemand1"] = "Fair"
df.loc[df['rentdemand1'].str.contains("Good|good|Good |Very good|Good |God|good |T Good|Good single family|Good except in winter months on unpaved streets|Good (for summer months only)|Owner area good|good for cheaper units|", na=False), "rentdemand1"] = "Good"
df.loc[df['rentdemand1'].str.contains("Owner occupancy|Owner Area|owner occupancy|Owner occupied|Owner area|Owner Occupied|Owner Occupancy|", na=False), "rentdemand1"] = "Owner Occupied"
df.loc[df['rentdemand1'].str.contains("Only one rent house in the area|Only two rent houses, both occupied|Few units|Limited number of units|Few|few rentals|Few rental units|Practically no rentals in this area|slight|Slight|pratically none|practically none|", na=False), "rentdemand1"] = "Few"
df.loc[df['rentdemand1'].str.contains("Excellent|Xcellent|excellent|", na=False), "rentdemand1"] = "Excellent"
```
